### Columnas para no tener en cuenta

```python
cols_except = ['NomEmpleadoRad','Idcontacto','hechoPor','nombreHechoPor',
               'usuarioNT responsable','AñoRadica','MesRadica','DiaRadica',
               'AñoEstado','MesEstado','DiaEstado']
cols = [col for col in df.columns.values if col not in cols_except]
```

### Modificación del tipo de dato para las columnas

```python
cols_dict_dtypes = {
    'NomGrupoRadicador':pd.StringDtype(),
    'fechaRadicacion':'datetime64[ns]',
    'SitioRecepcion':pd.StringDtype(),
    'MedioRecepcion':'category',
    'Prioridad':'category',
    'SucursalClasificacion':'category',
    'Categoria':'category',
    'Area':'category',
    'Causa':'category',
    #'Solucionado':'bool',
    'IdTipoContacto':pd.Int32Dtype(),
    'Oportunidad':'category',
    'GrupoResponsableActual':pd.StringDtype(),
    'SucursalResponsableActual':pd.StringDtype(),
    'TipoContacto':'category',
    'IpsUsuario':pd.StringDtype(),
    'Estado':'category',
    'fechaEstado':'datetime64[ns]',
    'Cod_IPSOriginaQueja':'object',
    'Nom_IPSOriginaQueja':pd.StringDtype(),
    'SubRed':'category',
    'SitioOrigen':'category',
    'comentariosCliente1':pd.StringDtype(),
    'comentariosCliente2':pd.StringDtype(),
    'ResponsableInicial':'category',
    'GrupoResponsableInicial':'category',
    'nombCategoriaInicial':'category',
    'nombAreaInicial':'category',
    'nombCausaInicial':'category',
    'histPQR':pd.Int32Dtype(),
    'nombreCoordinacionResponsableActual':'category',
    'tipoPQR':'category'
}
```

### Carga del archivo de acuerdo a los hallazgos

```python
df = pd.read_excel(os.path.join(DATASET_PATH,FILE_NAME),
                        header = 0,
                        dtype = cols_dict_dtypes,
                        usecols = cols,
                        true_values = ['SI'],
                        false_values = ['NO'])
```

### Modificaciones adicionales

```python
df['comentariosCliente2'].fillna('', inplace = True) # únicamente el segundo comentario
```

```python
df.rename({'comentariosCliente1':'comentariosCliente'}, axis = 'columns', inplace = True)
df['comentariosCliente'] = df['comentariosCliente'] + df['comentariosCliente2']
df.drop('comentariosCliente2', axis = 'columns', inplace = True, errors = 'ignore')
```

### Perfilado

```python
profile = df.profile_report()
profile = df.profile_report(title='Perfilado de dataset de PQRs',
                            infer_dtypes = False,
                            samples = {'head':0, 'tail':0},
                            interactions = {'continuous': False},
                            correlations = {'pearson':{'calculate':False}, 'spearman':{'calculate':False}, 'kendall':{'calculate':False}},
                            explorative=True)
profile.to_file(output_file="profiling_pqr.html")
```
